```python
import pandas as pd
df=pd.read_csv('youtube_unpopular1.csv', index_col=0)
df.reset_index(drop=True, inplace=True)
df.head(10)
```

```python
df.info()
```

```python
youtube_df1 = pd.read_csv('youtube_df1.csv', index_col=0)
youtube_df2 = pd.read_csv('youtube_df2.csv', index_col=0)
youtube_df3 = pd.read_csv('youtube_df3.csv', index_col=0)
youtube_df4 = pd.read_csv('youtube_df5.csv', index_col=0)
youtube_df5 = pd.read_csv('youtube_df4.csv', index_col=0)
youtube_df6 = pd.read_csv('youtube_df6.csv', index_col=0)
youtube_df7 = pd.read_csv('youtube_df7.csv', index_col=0)
youtube_df8 = pd.read_csv('youtube_df8.csv', index_col=0)
```

```python
print(youtube_df1.shape)
print(youtube_df2.shape)
print(youtube_df3.shape)
print(youtube_df4.shape)
print(youtube_df6.shape)
print(youtube_df7.shape)
print(youtube_df8.shape)
```

```python
dfdf=pd.concat([youtube_df1, youtube_df2, youtube_df3, youtube_df4, youtube_df5, youtube_df6, youtube_df7, youtube_df8], ignore_index=True)
dfdf
```

```python
df=pd.concat([df,dfdf],axis=1)
df.head()
```

```python
df[(df["likes"]=="0") | (df["unlikes"]=="0") | (df["comments"]=="0")]
```

```python
# 혹시 비공개 상태가 아니라 진짜 오류로 잘못 입력된 데이터 있나 해서..ㅎ
print(" or", df[(df["likes"]=="0") | (df["unlikes"]=="0")].shape)
print("and", df[(df["likes"]=="0") & (df["unlikes"]=="0")].shape)
print("not", df[(df["likes"]!="0") | (df["unlikes"]!="0")].shape)
```

```python
rate_dis=df[(df["likes"]=="0") | (df["unlikes"]=="0")]
rate_no=df[(df["likes"]!="0") | (df["unlikes"]!="0")]
rate_dis["ratings_disabled"]=1
rate_no["ratings_disabled"]=0
df=pd.concat([rate_no, rate_dis],axis=0)
df.shape
```

```python
print("댓글 비공개", df[df["comments"]=="0"].shape)
print("댓글 공개", df[df["comments"]!="0"].shape)
```

```python
comm_dis= df[df["comments"]=="0"]
comm_no=df[df["comments"]!="0"]
comm_dis["comments_disabled"]=1
comm_no["comments_disabled"]=0
df=pd.concat([comm_no, comm_dis],axis=0)
df.shape
```

```python
df[df["likes"]=="1개"]
```

```python
for i in range(len(df)):
    df["likes"].iloc[i]=df["likes"].iloc[i].split("개")[0]
    df["unlikes"].iloc[i]=df["likes"].iloc[i].split("개")[0]
    df["comments"].iloc[i]=df["likes"].iloc[i].split("개")[0]
df.head()
```

```python
df["likes"].replace('없음', 0)
df["unlikes"].replace('없음', 0)
df["comments"].replace('없음', 0)
```

```python
df["tags_split"]=0
df["tags_len"]=0
for i in range(len(df)):
    try:
        df["tags_split"].iloc[i] = df["tags"].iloc[i].split(", ")
        df["tags_len"].iloc[i] = len(df["tags_split"].iloc[i])
    except TypeError:
        df["tags_split"].iloc[i] = 0
        df["tags_len"].iloc[i] = 0
del df["tags"]
```

```python
df=df.rename(columns={'unlikes' : 'dislikes', 'comments' : 'comment_count', 'publishedAt' : 'publishedAt_date'})
df.head(3)
```

```python
df=df[['video_id', 'video_title', 'channel_id', 'channel_title', 'category', 'publishedAt_date', 'publishedAt_week',
       'tags_split', 'tags_len', 'view_count', 'play_time', 'likes', 'dislikes', 'comment_count', 'comments_disabled',
       'ratings_disabled','description']]
df.head(3)
```

```python
df.to_csv("youtube_unpopular.csv")
```
