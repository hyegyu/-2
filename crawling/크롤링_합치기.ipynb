```python
# í˜¹ì‹œ ë¹„ê³µê°œ ìƒíƒœê°€ ì•„ë‹ˆë¼ ì§„ì§œ ì˜¤ë¥˜ë¡œ ì˜ëª» ì…ë ¥ëœ ë°ì´í„° ìˆë‚˜ í•´ì„œ..ã…
print(" or", df[(df["likes"]=="0") | (df["unlikes"]=="0")].shape)
print("and", df[(df["likes"]=="0") & (df["unlikes"]=="0")].shape)
print("not", df[(df["likes"]!="0") | (df["unlikes"]!="0")].shape)
```

```python
rate_dis=df[(df["likes"]=="0") | (df["unlikes"]=="0")]
rate_no=df[(df["likes"]!="0") | (df["unlikes"]!="0")]
rate_dis["ratings_disabled"]=1
rate_no["ratings_disabled"]=0
df=pd.concat([rate_no, rate_dis],axis=0)
df.shape
```

```python
print("ëŒ“ê¸€ ë¹„ê³µê°œ", df[df["comments"]=="0"].shape)
print("ëŒ“ê¸€ ê³µê°œ", df[df["comments"]!="0"].shape)
```

```python
comm_dis= df[df["comments"]=="0"]
comm_no=df[df["comments"]!="0"]
comm_dis["comments_disabled"]=1
comm_no["comments_disabled"]=0
df=pd.concat([comm_no, comm_dis],axis=0)
df.shape
```

```python
df[df["likes"]=="1ê°œ"]
```

```python
for i in range(len(df)):
    df["likes"].iloc[i]=df["likes"].iloc[i].split("ê°œ")[0]
    df["unlikes"].iloc[i]=df["likes"].iloc[i].split("ê°œ")[0]
    df["comments"].iloc[i]=df["likes"].iloc[i].split("ê°œ")[0]
df.head()
```

```python
df["likes"].replace('ì—†ìŒ', 0)
df["unlikes"].replace('ì—†ìŒ', 0)
df["comments"].replace('ì—†ìŒ', 0)
```

```python
df["tags_split"]=0
df["tags_len"]=0
for i in range(len(df)):
    try:
        df["tags_split"].iloc[i] = df["tags"].iloc[i].split(", ")
        df["tags_len"].iloc[i] = len(df["tags_split"].iloc[i])
    except TypeError:
        df["tags_split"].iloc[i] = 0
        df["tags_len"].iloc[i] = 0
del df["tags"]
```

```python
df=df.rename(columns={'unlikes' : 'dislikes', 'comments' : 'comment_count', 'publishedAt' : 'publishedAt_date'})
df.head(3)
```

```python
df=df[['video_id', 'video_title', 'channel_id', 'channel_title', 'category', 'publishedAt_date', 'publishedAt_week',
       'tags_split', 'tags_len', 'view_count', 'play_time', 'likes', 'dislikes', 'comment_count', 'comments_disabled',
       'ratings_disabled','description']]
df.head(3)
```
