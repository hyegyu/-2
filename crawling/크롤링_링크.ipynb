```python
import pandas as pd
import numpy as np
df = pd.read_csv('youtube_kr.csv', index_col=0)
df.head(5)
```

```python
df.info()
```

```python
df_first=df.drop_duplicates(['video_id'], keep = 'first')
df_first=df_first.reset_index()
del df_first["index"]

df_first["tags_list"]=0
for i in range(len(df_first)):
    n = len(df_first["tags_split"][i])
    df_first["tags_list"][i] = df_first["tags_split"][i][2 : (n-2)].split("', '")
```

```python
word=[]
for i in range(len(df_first)):
    for j in range(len(df_first['tags_list'].iloc[i])):
        word.append(df_first['tags_list'].iloc[i][j])
len(word)
```

```python
while '[None]' in word:
    word.remove('[None]')
len(word)
```

```python
from collections import Counter
counter = Counter(word)
most_common = counter.most_common(2000)
```

```python
most_common[999][0]
```

```python
#딕셔너리 형태라서 리스트 형태로 전환
word_query=[]
for i in range(0, 2000):
    word_query.append(most_common[i][0] + ", year")
word_query
```

```python
# 실수로 초기화해서 이 부분 결과가 지워짐
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
df = pd.DataFrame(index=range(0,1), columns=["video_title", "url"])

for i in range(2000):
    url = 'https://www.youtube.com/results?search_query={}'.format(word_query[i])
    driver = webdriver.Chrome('./chromedriver.exe')
    driver.get(url)
    soup = bs(driver.page_source, 'html.parser')
    driver.close()

    name = soup.select('a#video-title')
    video_url = soup.select('a#video-title')

    name_list = []
    url_list = []

    for i in range(len(name)):
        name_list.append(name[i].text.strip())
    for i in video_url:
        url_list.append('{}{}'.format('https://www.youtube.com',i.get('href')))

    youtubeDic = {
        'video_title': name_list,
        'url': url_list
    }
    df_to_insert = pd.DataFrame(youtubeDic)
    df = df.append(df_to_insert, ignore_index=True)
```
