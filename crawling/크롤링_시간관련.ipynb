```python
import pandas as pd
import numpy as np
df = pd.read_csv('youtube_unpopular1.csv', index_col=0)
df.head(5)
```

```python
df["description"]=df["description"].fillna(0)
df["description"][df["description"] != 0] = 1
df.head()
```

```python
df['publishedAt']=pd.to_datetime(df['publishedAt'])
df["publishedAt_week"]=df["publishedAt"].dt.day_name()
df.head()
```

```python
import json

id_to_category = {}

with open('KR_category_id.json', 'r') as f:
    data = json.load(f)
    for category in data['items']:
        id_to_category[category['id']] = category['snippet']['title']
```

```python
id_to_category={v:k for k,v in id_to_category.items()}
```

```python
df=df.replace({"category" : id_to_category})
```

```python
count={}
lists = df["channel_id"]
for i in lists:
    try: count[i] += 1
    except: count[i]=1
print(count)
```

```python
import operator
sorted(count.items(), key=operator.itemgetter(1), reverse=True)
```

```python
from pytube import YouTube

play_time=[]
for i in range(len(df)):
    url = "https://www.youtube.com/watch?v=" + df["video_id"].iloc[i]
    yt = YouTube(url)
    try:
        play_time.append(yt.length)
    except TypeError:
        play_time.append(0)
play_time
```

```python
df.insert(10,"play_time",play_time,True)
df
```

```python
df.to_csv("youtube_unpopular1.csv")
```
