```python
#ì˜ë¬¸ìœ¼ë¡œ ë³€í™˜ë˜ì–´ ìˆ˜ì§‘ëœ ë°ì´í„° ì¡´ì¬
link_df=link_df.drop_duplicates(subset=["video_id"], ignore_index=True)
link_df
```

```python
df1=origin["video_id"].drop_duplicates()
df2=link_df["video_id"]

dfdf=pd.concat([df2, df1])
dfdf
```

```python
del_index=dfdf[dfdf.duplicated(keep='last')].index
```

```python
link_df=link_df.drop(index=del_index, axis=0)
link_df=link_df.reset_index(drop = True)
link_df
```

```python
link_df["url"].iloc[30]
```

```python
import requests
from bs4 import BeautifulSoup

channel_id=[]
channel_title=[]
category=[]
publishedAt=[]
tags=[]
view_count=[]
description=[]
for i in range(len(link_df)):
    url=link_df["url"].iloc[i]

    headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
    data = requests.get(url,headers=headers)
    
    soup = BeautifulSoup(data.text, 'html.parser')
    try:
        channel_id.append(soup.select_one('meta[itemprop="channelId"]')['content'])
    except:
        channel_id.append(0)
        
    try:
        channel_title.append(soup.select_one('link[itemprop="name"]')['content'])
    except:
        channel_title.append(0)
        
    try:     
        category.append(soup.select_one('meta[itemprop="genre"]')['content'])
    except:
        category.append(0)
        
    try:
        publishedAt.append(soup.select_one('meta[itemprop="datePublished"]')['content'])
    except:
        publishedAt.append(0)
        
    try:
        tags.append(soup.select_one('meta[name="keywords"]')['content'])
    except:
        tags.append(0)    
        
    try:
        view_count.append(soup.select_one('meta[itemprop="interactionCount"]')['content'])
    except:
        view_count.append(0)  
        
    try:
        description.append(soup.select_one('meta[itemprop="description"]')['content'])
    except:
        description.append(0)            
```

```python
data = {'channel_id': channel_id, 'channel_title': channel_title,'category': category, 'publishedAt':publishedAt,
        'tags': tags, 'view_count': view_count,'description': description}
df = pd.DataFrame(data)
df
```

```python
df_merge=pd.concat([link_df, df], axis=1)
df_merge
```

```python
df_merge['publishedAt']=pd.to_datetime(df_merge['publishedAt'])
df_merge.info()
```

```python
df_merge=df_merge[df_merge["view_count"]!=0]
df_merge=df_merge.reset_index(drop = True)
df_merge
```

```python
df_filered = df_merge[df_merge['publishedAt'].between('2020-08-12', '2021-09-14')]
df_filered
```
