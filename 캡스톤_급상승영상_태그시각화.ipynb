## 전체 태그 워드 클라우드

```python
df_up['tags_split']
```

```python
words=[]
for i in range(len(df_up)):
    word_list=df_up['tags_split'].iloc[i]
    if word_list=="['[None]']":
        continue
    tag=word_list[2:len(word_list[1])-3].split("', '")    
    for j in range(len(tag)):
        
        words.append(tag[j])
len(words)
```

```python
from collections import Counter 
count = Counter(words) 
com_count = count.most_common(100)
```

```python
com_count[0:10]
```

```python
from os import path
from PIL import ImageFont
from PIL import Image 
from wordcloud import WordCloud

mask=np.array(Image.open("유튜브 마스킹 그림.png")) 
#검은 부분, 흰색부분으로만 된 사진을 배열형식으로 받음. 근데 왜인지 모르겠지만 0으로 나와서 검은색인 255로 바꿔주기
for i in range(len(mask)):
    for j in range(len(mask[i])):
        if mask[i][j]==0:
            mask[i][j] = 255
wc = WordCloud(font_path=font_path, background_color='white', mask=mask,
               width=1000, height=800, max_font_size=300)
```

```python
wc.generate_from_frequencies(dict(com_count))
plt.figure(figsize=(15, 10))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
