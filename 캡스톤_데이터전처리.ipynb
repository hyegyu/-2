```python
#pip install git+git://github.com/nficano/pytube <- git ì„¤ì¹˜í•˜ê³  ì‹œìŠ¤í…œ ë³€ìˆ˜ path ì„¤ì •í•˜ê³  ì§„í–‰í•˜ë©´ í•´ê²° >_0
#ì²˜ìŒì— ë¬´ì‹í•˜ê²Œ ì›ë³¸ë°ì´í„°ì— ë„£ì—ˆëŠ”ë° êµ‰ì¥íˆ ì˜¤ë˜ê±¸ë ¤ì„œ(ì•„ë§ˆ 1-2ì‹œê°„?) ë‹¤ì‹œëŠ” í•˜ê³  ì‹¶ì§€ ì•ŠìŒ
#ê·¸ë˜ì„œ ì˜ìƒì´ ì¤‘ë³µì¸ì§€ í™•ì¸í–ˆê³  ì¤‘ë³µëœ ì˜ìƒ ë¦¬ìŠ¤íŠ¸ì—ì„œ ëŒë¦¼. ë°ì´í„° ì–‘ì´ 1/7ë¨.
from pytube import YouTube

play_time=[]
for i in range(len(count)):
    url = "https://www.youtube.com/watch?v=" + df2["index"][i]
    yt = YouTube(url)
    try:
        play_time.append(yt.length)
    except TypeError:
        play_time.append(0)
    print(i, "ë²ˆì§¸ ì‹œê°„ :", play_time[i])
play_time
```

```python
len(play_time)
```

```python
df2=df2.rename(columns = {0 : "duration"})
df2["play_time"]=play_time
df2.head(3)
```

```python
# ê¸°ì¡´ ë°ì´í„°ì— ìƒˆë¡œìš´ ë°ì´í„° ì¶”ê°€
import operator
df = pd.merge(df,df2, how='left', left_on='video_id', right_on='index')
del df["index"]
df.head(5)
```

## booleaní˜• ì²˜ë¦¬

```python
df["comments_disabled"]= df["comments_disabled"].replace([True,False], [1, 0])
df["ratings_disabled"]= df["ratings_disabled"].replace([True,False], [1, 0])
df.head(3)
```

## íƒœê·¸ ì²˜ë¦¬

```python
df["tags_split"]=0
df["tags_len"]=0
for i in range(len(df)):
    try:
        df["tags_split"].iloc[i] = df["tags"].iloc[i].split("|")
        df["tags_len"].iloc[i] = len(df["tags_split"].iloc[i])
    except TypeError:
        df["tags_split"].iloc[i] = 0
        df["tags_len"].iloc[i] = 0
df.head(3)

del df["tags"]
```
