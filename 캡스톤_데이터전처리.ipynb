## 날짜형 데이터 변경

```python
df[["publishedAt", "trending_date"]]
```

```python
df["publishedAt_date"]=df["publishedAt"].str.split('T').str[0]
df["publishedAt_time"]=df["publishedAt"].str.split('T').str[1]
df["publishedAt_time"]=df["publishedAt_time"].str.split('Z').str[0]
df["trending_date"]=df["trending_date"].str.split('T').str[0]
```

```python
df[["publishedAt_date", "publishedAt_time", "trending_date"]]
```

```python
df.info()
```

```python
df[["trending_date", "publishedAt", "publishedAt_date", "publishedAt_time"]]
```

```python
del df['publishedAt']
df['trending_date']=pd.to_datetime(df['trending_date'], format="%Y-%m-%d %H:%M:%S")
df["publishedAt_date"]=pd.to_datetime(df['publishedAt_date'], format="%Y-%m-%d %H:%M:%S")
df["publishedAt_time"]=pd.to_datetime(df['publishedAt_time'], format="%H:%M:%S")
```

```python
#요일 추가
import datetime
df["publishedAt_week"]=df["publishedAt_date"].dt.day_name()
df["trending_week"]=df["trending_date"].dt.day_name()
```

```python
#유일하게 결측값이 있는 데이터라 없는 건 - 0
df["description"]=df["description"].fillna(0)
```

```python
#설명이 뭐라도 있는거면 1
df["description"][df["description"] != 0] = 1
```

## 재생시간 크롤링

```python
df1=df.drop_duplicates(["video_id"], keep = 'first')
df1.info()
```

```python
#pip install git+git://github.com/nficano/pytube <- git 설치하고 시스템 변수 path 설정하고 진행하면 해결 >_0
#처음에 무식하게 원본데이터에 넣었는데 굉장히 오래걸려서(아마 1-2시간?) 다시는 하고 싶지 않음
#그래서 영상이 중복인지 확인했고 중복된 영상 리스트에서 돌림. 데이터 양이 1/7됨.
from pytube import YouTube

play_time=[]
for i in range(len(count)):
    url = "https://www.youtube.com/watch?v=" + df2["index"][i]
    yt = YouTube(url)
    try:
        play_time.append(yt.length)
    except TypeError:
        play_time.append(0)
    print(i, "번째 시간 :", play_time[i])
play_time
```

```python
len(play_time)
```

```python
df2=df2.rename(columns = {0 : "duration"})
df2["play_time"]=play_time
df2.head(3)
```

```python
# 기존 데이터에 새로운 데이터 추가
import operator
df = pd.merge(df,df2, how='left', left_on='video_id', right_on='index')
del df["index"]
df.head(5)
```

## boolean형 처리

```python
df["comments_disabled"]= df["comments_disabled"].replace([True,False], [1, 0])
df["ratings_disabled"]= df["ratings_disabled"].replace([True,False], [1, 0])
df.head(3)
```

## 태그 처리

```python
df["tags_split"]=0
df["tags_len"]=0
for i in range(len(df)):
    try:
        df["tags_split"].iloc[i] = df["tags"].iloc[i].split("|")
        df["tags_len"].iloc[i] = len(df["tags_split"].iloc[i])
    except TypeError:
        df["tags_split"].iloc[i] = 0
        df["tags_len"].iloc[i] = 0
df.head(3)

del df["tags"]
```

```python
df.info()
```

```python
#보기편한 이름으로 변환
df = df.rename(columns={'title': 'video_title', 'channelId': 'channel_id', 'channelTitle': 'channel_title', 'categoryId': 'category'})
df.info()
```
